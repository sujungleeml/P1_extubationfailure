```python
# 함수: 중복된 이벤트 확인
def is_duplicated_event(row):
    # 현재 row의 hadm_id와 intubationtime을 가져옵니다.
    current_hadm_id = row['hadm_id']
    current_intubation_time = row['intubationtime']
    
    # 이전 row의 hadm_id와 intubationtime을 가져옵니다.
    previous_hadm_id = row['prev_hadm_id']
    previous_intubation_time = row['prev_intubationtime']
    
    # 현재 row와 이전 row 사이의 intubationtime의 시간 차이를 계산합니다.
    if pd.notnull(previous_intubation_time):
        time_difference = current_intubation_time - previous_intubation_time
    else:
        time_difference = None

    # 현재 row와 이전 row의 hadm_id가 동일하고 intubationtime의 차이가 1시간 이내인 경우 "duplicated", 그렇지 않으면 "not duplicated" 반환
    if current_hadm_id == previous_hadm_id and (time_difference is not None) and (time_difference <= pd.Timedelta(hours=1)):
        return 'duplicated'
    return 'not duplicated'

# 이전 row의 hadm_id와 intubationtime을 새로운 컬럼으로 추가합니다.
intubation1['prev_hadm_id'] = intubation1['hadm_id'].shift(1)
intubation1['prev_intubationtime'] = intubation1['intubationtime'].shift(1)

# 새로운 'dup' 컬럼에 is_duplicated_event 함수를 적용하여 결과를 저장합니다.
intubation1['dup'] = intubation1.apply(is_duplicated_event, axis=1)

# 'prev_hadm_id' 및 'prev_intubationtime' 컬럼 삭제
intubation1.drop(columns=['prev_hadm_id', 'prev_intubationtime'], inplace=True)
```
