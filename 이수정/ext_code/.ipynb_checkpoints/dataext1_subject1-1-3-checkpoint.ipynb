```python
df = error_rows_in4_2 

def process_data(group_df):
    # intubationtime과 extubationtime에 따라 데이터 추출
    int_df = group_df[['subject_id', 'hadm_id','intubationtime', 'admittime', 'int_itemid', 'int_weight','intcount', 'intseq']].drop_duplicates(subset=['intubationtime']).reset_index(drop=True)
    ext_df = group_df[['subject_id', 'hadm_id','extubationtime', 'ext_itemid', 'ext_weight', 'extubationcause', 'intubationtoextubationtime', 'dischtime', 'deathtime', 'intcount', 'intseq']].drop_duplicates(subset=['extubationtime']).reset_index(drop=True)

    # int_df와 ext_df의 hadm_id 개수 비교
    intcount = max(len(int_df), len(ext_df))

    # int_df에 intseq 부여
    int_df = int_df.sort_values(by='intubationtime', ascending=True)
    int_df['intseq'] = range(1, len(int_df) + 1)

    # ext_df에 intseq 부여
    ext_df = ext_df.sort_values(by='extubationtime', ascending=True)
    ext_df['intseq'] = range(1, len(ext_df) + 1)
    ext_df['intseq'] = ext_df['intseq'] + abs(len(int_df)-len(ext_df))

    # 두 데이터프레임 병합
    merged_df = pd.merge(int_df, ext_df, on=['subject_id', 'hadm_id','intcount', 'intseq'], how='outer')
    
    return merged_df

grouped = df.groupby('hadm_id')

result_dfs = []
for _, group_df in grouped:
    result_dfs.append(process_data(group_df))

df2 = pd.concat(result_dfs)
df2 
#intseq 부여
```

```python
def fill_na(group):
    for idx, row in group.iterrows():
        next_row = group.iloc[idx + 1] if idx + 1 < len(group) else None
        if next_row is not None:
            group['admittime'] = group['admittime'].fillna(method='ffill')
            group['extubationtime'] = group['extubationtime'].fillna(method='bfill')
            group['extubationcause'] = group['extubationcause'].fillna(method='bfill')
            group['intubationtoextubationtime'] = group['intubationtoextubationtime'].fillna(method='bfill')
            return group
        else:
            return group

# hadm_id 별로 그룹화 후, 각 그룹에 대해 fill_na 함수를 적용
df1 = df.groupby('hadm_id').apply(fill_na).reset_index(drop=True)

df1['hadm_id'].count()

```

#### 함수정의

```python
#함수 수정이 필요함

def process_group4_updated(group):
    group = group.copy()
    
    for idx in group.index:
        # 현재 row와 다음 row를 비교하기 위해 다음 row의 정보를 가져온다.
        prev_idx = group.index[group.index < idx].max()  # 이전 행의 인덱스 가져오기
        next_idx = group.index[group.index > idx].min()  # 다음 행의 인덱스 가져오기

        prev_row = group.loc[prev_idx] if pd.notnull(prev_idx) else None
        next_row = group.loc[next_idx] if pd.notnull(next_idx) else None

        if next_row is not None and prev_row is not None:
            if group.loc[idx, 'intubationtime'] > group.loc[idx, 'extubationtime'] and group.loc[idx, 'extubationtime'] == prev_row['extubationtime']:
                group.loc[idx, 'extubationtime'] = next_row['extubationtime']
                group.loc[idx, 'extubationcause'] = next_row['extubationcause']

            if group.loc[idx, 'extubationtime'] == prev_row['extubationtime']:
                if next_row is not None and group.loc[idx, 'extubationtime'] < next_row['extubationtime']:
                    group.loc[idx, 'extubationtime'] = next_row['extubationtime']
            
            if pd.notnull(group.loc[next_idx, 'intubationtime']):
                        if group.loc[next_idx, 'intubationtime'] > next_row['extubationtime']:
                            group.loc[idx, 'extubationtime'] = next_row['extubationtime']
                        elif group.loc[next_idx, 'intubationtime'] < next_row['extubationtime']:
                            group.loc[idx, 'extubationtime'] = next_row['intubationtime']
                            group.loc[idx, 'extubationcause'] = 'extubationtime replaced'

        if next_row is None and prev_row is not None:
            if group.loc[idx, 'intubationtime'] > group.loc[idx, 'extubationtime'] and group.loc[idx, 'extubationtime'] == prev_row['extubationtime']:
                if next_row is not None and group.loc[idx, 'extubationtime'] < next_row['intubationtime']:
                    group.loc[idx, 'extubationtime'] = next_row['extubationtime']
                    group.loc[idx, 'extubationcause'] = next_row['extubationcause']
                if next_row is not None and group.loc[idx, 'extubationtime'] > next_row['intubationtime']:
                    group.loc[idx, 'extubationtime'] = next_row['intubationtime']
                    group.loc[idx, 'extubationcause'] = 'extubationtime replaced'

        # 만약 현재 row의 intubationtime 보다 현재 row의 extubationtime 이 더 작을때.
        if next_row is not None and group.loc[idx, 'intubationtime'] > group.loc[idx, 'extubationtime']:
            
            # 만약 다음 줄의 intubationtime 이 NaT이면, 현재 줄의 intubationtime을 다음 줄의 intubationtime으로 대체한다.
            if pd.isnull(next_row['intubationtime']):
                group.loc[next_idx, 'intubationtime'] = group.loc[idx, 'intubationtime']

            if next_row is not None and prev_row is not None and group.loc[idx,'intubationtime'] > next_row['extubationtime']:
                
                i_source_values = group.loc[idx:, 'intubationtime'].dropna().values
                i_target_indexes = group.loc[idx+1:].index[:len(i_source_values)]
                
                if len(i_source_values) == len(i_target_indexes):
                    group.loc[i_target_indexes, 'intubationtime'] = i_source_values
                    group.loc[idx, 'intubationtime'] = group.loc[idx,'intubationtime']
                    group.loc[idx, 'extubationcause'] = 'intubationtime replaced'
                    group.loc[idx, 'intubationtime'] = group.loc[prev_idx, 'extubationtime']

                else:
                    print("Length mismatch: source =", len(i_source_values), ", target =", len(i_target_indexes), ", idx =", idx)

        
        if next_row is None and prev_row is not None and group.loc[idx, 'extubationtime'] == prev_row['extubationtime']:
            group.loc[idx, 'extubationtime'] = pd.NaT
 
        if next_row is None and pd.isnull(group.loc[idx, 'extubationtime']):
            if prev_row is not None and group.loc[idx, 'intubationtime'] == prev_row['intubationtime']:
                group.loc[idx, 'intubationtime'] = pd.NaT

    return group
```

## Reintubationtime 처리

```python
# extubationto reintubationtime 을 계산해야 하는데,
# intcount 가 1인 경우에는 계산할 필요가 없음 reintubationtime = 0 으로 처리
# intcount 가 2 이상인 경우부터 계산
# reintubationtime은 다음행의 intubationtime - 현재 행의 extubationtime  현재행에 결과 도출
# 마지막 행은 reintubationtime 이 없는 것이 정상. 그래서 마지막 행은 reintubationtime = 0 으로 처리
```

```python
def calculate_reintubation_time(group):
    # 그룹을 인덱스 순서대로 정렬 (중요함, 인덱스가 비순차적일 수 있으므로)
    group = group.sort_index()
    
    # `reintubationtime` 컬럼을 0으로 초기화
    group['reintubationtime'] = 0
    
    # 마지막 행을 제외한 모든 행을 반복
    for i in range(len(group) - 1):
        current_row = group.iloc[i]
        next_row = group.iloc[i + 1]
        
        # `intcount`가 2 이상인 경우에만 계산
        if current_row['intcount'] >= 2:
            # 다음 행의 `intubationtime`에서 현재 행의 `extubationtime`을 뺀다
            group.at[current_row.name, 'reintubationtime'] = (next_row['intubationtime'] - current_row['extubationtime']).total_seconds() / 3600  # 초 단위로 계산한 다음 시간으로 변환
    
    # 마지막 행 처리
    # 마지막 행에 대해서는 `reintubationtime`을 0으로 설정 (이미 초기화 되었으므로 생략 가능)
    
    return group

# 이 함수를 각 그룹(예: `subject_id`별로 그룹화된 데이터프레임)에 적용합니다.
# 예: df.groupby('subject_id').apply(calculate_reintubation_time)
```

```python
# 데이터프레임이 'df'라고 가정하고, 'subject_id'로 그룹화한 다음 함수를 적용
reintubation1 = updated_df2.groupby('hadm_id').apply(calculate_reintubation_time).reset_index(drop=True)
```

```python
def process_reintubation1_updated(group):
    group = group.copy()
    
    for idx in group.index:
        # 이전 행과 다음 행의 인덱스를 가져옵니다.
        prev_idx = group.index[group.index < idx].max() if not group.index[group.index < idx].empty else None
        next_idx = group.index[group.index > idx].min() if not group.index[group.index > idx].empty else None

        prev_row = group.loc[prev_idx] if prev_idx is not None else None
        next_row = group.loc[next_idx] if next_idx is not None else None

        # 복사한 line인지 확인하고 처리합니다.
        if prev_row is not None and next_row is not None:
            if group.loc[idx, 'reintubationtime'] < 0 and next_row['reintubationtime'] >= 0:
                # 현재 row가 복사한 line일 경우
                if prev_row['intubationtime'] == group.loc[idx, 'intubationtime'] and prev_row['extubationtime'] == group.loc[idx, 'extubationtime']:
                    # 다음 row의 intubationtime이 현재 row의 extubationtime보다 작은 경우
                    if next_row['intubationtime'] < group.loc[idx, 'extubationtime']:
                        group.loc[next_idx, 'intubationtime'] = group.loc[idx, 'extubationtime']
                        group.loc[next_idx, 'extubationcause'] = 'intubation replaced'
                # 다음 row의 extubationtime이 현재 row의 intubationtime보다 작은 경우
                if next_row['extubationtime'] < group.loc[idx, 'intubationtime']:
                    group.loc[idx, 'extubationtime'] = next_row['intubationtime']
                    group.loc[idx, 'extubationcause'] = 'extubation replaced'
                # 현재 row의 extubationtime이 다음 row의 intubationtime보다 큰 경우
                if group.loc[idx, 'extubationtime'] > next_row['intubationtime']:
                    group.loc[prev_idx, 'extubationtime'] = next_row['intubationtime']
                    group.loc[idx, 'intubationtime'] = next_row['intubationtime']
                    group.loc[prev_idx, 'extubationcause'] = 'intubation replaced'

    return group
```

```python
def drop_lastrow(group):
    group = group.copy()
    rows_to_drop = []

    for idx in group.index:
        prev_idx = group.index[group.index < idx].max() if not group.index[group.index < idx].empty else None

        if prev_idx is not None:
            prev_row = group.loc[prev_idx]
            if prev_row['reintubationtime'] < 0:
                rows_to_drop.append(idx)

    group.drop(rows_to_drop, inplace=True)
    return group
```

```python
def group_time1(group):
    group = group.sort_index()
    
    for idx in group.index:
        # 현재 row와 다음 row를 비교하기 위해 다음 row의 정보를 가져온다.
        # 최대/최소 인덱스가 존재하지 않을 경우를 처리합니다.
        prev_idx = group.index[group.index < idx].max() if not group.index[group.index < idx].empty else None
        next_idx = group.index[group.index > idx].min() if not group.index[group.index > idx].empty else None

        prev_row = group.loc[prev_idx] if prev_idx is not None else None
        next_row = group.loc[next_idx] if next_idx is not None else None
        
        # 첫 번째 행을 처리합니다.
        if prev_row is None and pd.isnull(group.loc[idx, 'extubationtime']):
            if next_row is not None:
                if group.loc[idx, 'reintubationtime'] < 0:
                    # intubationtime 이 중복될 때
                    if group.loc[idx, 'intubationtime'] == next_row['intubationtime'] and group.loc[idx, 'extubationtime'] != next_row['extubationtime']:
                        group.loc[idx, 'intubationtime'] = group.loc[idx,'admittime']
                        group.loc[idx, 'extubationcause'] = 'intubation replaced'

                    # extubationtime 이 중복될 때
                    if next_row['intubationtime'] < group.loc[idx, 'extubationtime'] and group.loc[idx, 'extubationtime'] == next_row['extubationtime']:
                        group.loc[idx, 'extubationtime'] = next_row['intubationtime']
                        group.loc[idx, 'extubationcause'] = 'extubation replace'

    return group
```

```python
df2 = df1.groupby('hadm_id').apply(group_time1).reset_index(drop=True)
df2
```

```python
def calculate_reintubation_time(group):
    # 그룹을 인덱스 순서대로 정렬 (중요함, 인덱스가 비순차적일 수 있으므로)
    group = group.sort_index()
    
    # `reintubationtime` 컬럼을 0으로 초기화
    group['reintubationtime'] = 0
    
    # 마지막 행을 제외한 모든 행을 반복
    for i in range(len(group) - 1):
        current_row = group.iloc[i]
        next_row = group.iloc[i + 1]
        
        # `intcount`가 2 이상인 경우에만 계산
        if current_row['intcount'] >= 2:
            # 다음 행의 `intubationtime`에서 현재 행의 `extubationtime`을 뺀다
            group.at[current_row.name, 'reintubationtime'] = (next_row['intubationtime'] - current_row['extubationtime']).total_seconds() / 3600  # 초 단위로 계산한 다음 시간으로 변환
    
    # 마지막 행 처리
    # 마지막 행에 대해서는 `reintubationtime`을 0으로 설정 (이미 초기화 되었으므로 생략 가능)
    
    return group

# 이 함수를 각 그룹(예: `subject_id`별로 그룹화된 데이터프레임)에 적용합니다.
# 예: df.groupby('subject_id').apply(calculate_reintubation_time)
```

```python
# 데이터프레임이 'df'라고 가정하고, 'subject_id'로 그룹화한 다음 함수를 적용
df2 = df2.groupby('hadm_id').apply(calculate_reintubation_time).reset_index(drop=True)
```
