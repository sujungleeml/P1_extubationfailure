# 환자 정의

* 먼저 intubation -> extubation -> intubation(reintubation) 순으로 절차가 이루어 집니다. 여기서 intubation~extubation 이 적어도 24시간 이상 이루어진 환자를 찾습니다 (일시적 목적으로 사용한 환자 제거)
* 그다음, extubation 이후 48시간 이내에 다시 intubation 을 한 환자를 찾습니다. (reintubation) 이 환자들은 extubation failure 군으로 분류 될 것입니다.
* 그렇다면 extubation 이후 48시간을 초과하여 재 삽관(reintubation 한 환자는 extubation non-failure 로 분류 할 것입니다. sucess라고 단정지을 수 있는)

```python
def is_duplicated_event(row):
    current_hadm_id = row["hadm_id"]
    current_intubation_time = row["intubationtime"]

    previous_hadm_id = row["prev_hadm_id"]
    previous_intubation_time = row["prev_intubationtime"]

    if pd.notnull(previous_intubation_time):
        time_difference = current_intubation_time - previous_intubation_time
    else:
        time_difference = None

    if current_hadm_id == previous_hadm_id and (time_difference is not None) and (time_difference <= pd.Timedelta(hours=1)):
        return 'duplicated'
    return "not duplicated"

intubation1_dup["prev_hadm_id"] = intubation1_dup["hadm_id"].shift(1)
intubation1_dup["prev_intubationtime"] = intubation1_dup["intubationtime"].shift(1)

intubation1_dup["dup"] = intubation1_dup.apply(is_duplicated_event, axis=1)
intubation1_dup.drop(columns=["prev_hadm_id", "prev_intubationtime"], inplace=True)
```

```python
def label_outcomes(row):
    if pd.notnull(row["reintubationtime"]):
        return "reintubation"
    elif pd.notnull(row["deathtime"]):
        return "death"
    elif pd.notnull(row['dischtime']):
        return "discharged"
    else:
        return "unknown"
```

```python
def classify_outcome_step2_date(row):
    if row["outcome_step1"] == "reintubation":
        return "reintubation"
    elif row['outcome_step1'] == "death" and row['deathtime'].date() == row['extubationtime'].date():
        return "death"
    elif row["outcome_step1"] == "discharged" and row["dischtime"].date() == row['extubationtime'].date():
        return "discharged"
    else:
        return "unknown"
```
