### dignoses_icd

```python
start_time = time.time()

cursor = conn.cursor()
cursor.execute("select * from mimiciv_hosp.diagnoses_icd;")
rows = cursor.fetchall()
columns_name = [desc[0] for desc in cursor.description]
diagnoses_icd = pd.DataFrame(rows, columns=columns_name)

end_time = time.time()
wait_time = end_time - start_time
print("경과 시간 : ", wait_time)
```

```python
diagnoses_icd.head()
```

```python
# J44가 들어가면 copd 데이터임을 활용

diagnoses_copd = diagnoses_icd[diagnoses_icd["icd_code"].str.contains("J44")]
```
